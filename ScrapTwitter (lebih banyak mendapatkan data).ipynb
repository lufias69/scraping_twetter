```python
from urllib.request import urlopen
import pandas as pd
import codecs
from bs4 import BeautifulSoup
# from selenium import webdriver
# from selenium.webdriver.common.keys import Keys
import time
import datetime
```

```python
#downlod fila html yang akan diambil datanya

nama_file = "#JokowiPelanggarHAM.html"
f=codecs.open("html/"+nama_file, 'r', 'utf-8')
soup = BeautifulSoup(f, 'html.parser')
#stream
id_ = list()
twet_ = list()
stream = soup.find('div','stream')
stream_item = stream.find_all('li', 'js-stream-item stream-item stream-item')
for item in stream_item:
    nama = item.find('span','username u-dir u-textTruncate')
    id_.append(nama.get_text())
    twit = item.find('div', 'js-tweet-text-container')
    twet_.append(twit.get_text().encode("utf-8").decode("utf-8").rstrip('\r\n').lstrip().replace('\n',' '))
dictionary = {
    'username':id_,
    'twet':twet_
}
df1 = pd.DataFrame.from_dict(dictionary)
df1.to_excel("hasil/"+nama_file+".xlsx") 
df1
```
