```python
from urllib.request import urlopen
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import datetime
now = datetime.datetime.now()
sekarang = now.hour, now.minute, now.second
```

```python
hashtag  = "BelaKedaulatanRakyat"
driver = webdriver.Chrome(executable_path=r"driver/chromedriver.exe")
#options.add_argument(“--lang= iddelapan8*”)
driver.get("https://twitter.com/hashtag/"+hashtag+'?src=tren')
elm = driver.find_element_by_tag_name('html')
for i in range(50):
    elm.send_keys(Keys.END)
    time.sleep(3)
    if i > 10 and i%5==0:
        elm.send_keys(Keys.HOME)
        time.sleep(2)
        
        
thepage = driver.page_source
soup = BeautifulSoup(thepage, "lxml")

name = []
retwit = []
twit =[]
tanggal = []

for tweets in soup.findAll('div', {"class":"content"}):
    #print (nama.replace("|","").replace("Twitter",""))
    if (tweets.find('p',{"class":"TweetTextSize TweetTextSize--normal js-tweet-text tweet-text"})):
        nama = soup.find('b', {"class":"u-linkComplex-target"}).text
        nm = nama.replace("|","").replace("Twitter","")
        name.append(nm)
    
        if tweets.find('span', {"class":"username u-dir u-textTruncate"}):
            #print (tweets.find('span', {"class":"username u-dir u-textTruncate"}).text)
            retwit.append(tweets.find('span', {"class":"username u-dir u-textTruncate"}).text.replace("@",""))
        
        #print (tweets.find('p',{"class":"TweetTextSize TweetTextSize--normal js-tweet-text tweet-text"}).text)
        twit.append(tweets.find('p',{"class":"TweetTextSize TweetTextSize--normal js-tweet-text tweet-text"}).text)
        #print (tweets.find('p', {"class":"TweetTextSize TweetTextSize--normal js-tweet-text tweet-text"}).text)
        ds = tweets.find('small',{"class":"time"})
        #print (ds)
        teks = str(ds)
        cari = 'title='
        deli = '"'


        tangal = []
        cek = False
        pc = len(cari)
        q = 0
        h = 0
        apa =0
        j = 0
        
        for i in teks:
            kumpul = []
            if cek == False:
                q=h
                for j in cari:
                    kumpul.append(teks[q])
                    q = q+1   
                kumpul2 = ''.join(kumpul)
                h = h+1
                j =+1
                if kumpul2 == cari:
            
                    cek = True
                    q = q+1
                    j =+1
            else:
        
                if teks[q] != deli:
                    tangal.append(teks[q])
            
                    q = q+1
                    apa = apa+1
                else:
                    break
            tangal2 = ''.join(tangal)
        #print (tangal2)
        tanggal.append(tangal2)
       # print("------------------------------------")
    else:
        pass
    
    tweet_dict = {"nama":name,"retwet_name":retwit, "twit":twit, "tanggal":tanggal}
    df = pd.DataFrame(tweet_dict, columns = ["nama","retwet_name","twit","tanggal"] )
df

writer =pd.ExcelWriter(hashtag+'.xlsx')
df.to_excel(writer,'Sheet1')
writer.save()
```
